```python
import os

import numpy as np
import pandas as pd
import folium
from folium import plugins

print(folium.__version__)
```

```python
import pandas as pd
```

```python
#bring in the 6 packs
train=pd.read_csv("C:/Users/GDLSYSTEM/2890/df_day_all.csv")#../input/nyc-taxi-trip-duration/
train.drop(['Unnamed: 0', 'TRGT_SENSOR_KIND','MSG_ID','LOC_ACCRCY'], axis='columns', inplace=True)
train.head(3)
```

```python
#Extract the month column from pickup datetime variable and take subset of data
train['RECV_DT'] = pd.to_datetime(train.RECV_DT)
train['RECV_DT'] = train['RECV_DT'].dt.day
train.tail(3)
```

```python
heat_df =train.sample(n=1000000)
```

```python
heat_df = heat_df[['LAT_VAL', 'LON_VAL','RECV_DT']]
```

```python
heat_df['LAT_VAL'] = heat_df['LAT_VAL'].astype(float)
heat_df['LON_VAL'] = heat_df['LON_VAL'].astype(float)
```

```python
heat_df = heat_df.dropna(axis=0)
```

```python
heat_df['Weight'] = heat_df['RECV_DT']
#heat_df['Weight'] = heat_df['Weight'].astype(str)
heat_df['Weight'] = heat_df['Weight'].astype(float)
heat_df
```

```python
heat_df = heat_df.dropna(axis=0, subset=['LAT_VAL','LON_VAL', 'Weight'])
heat_df
```

```python
heatmap = folium.Map(location=[36.99543,126.7693785],tiles= "Stamen Terrain",
                    zoom_start = 10)
```

```python
heat_data = [[[row['LAT_VAL'],row['LON_VAL']] 
                for index, row in heat_df[heat_df['Weight'] == i].iterrows()] 
                 for i in range(0,19)]
```

```python
hm = plugins.HeatMapWithTime(heat_data,auto_play=True,max_opacity=0.8)
hm.add_to(heatmap)
```
